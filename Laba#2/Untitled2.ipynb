### Обработка пропусков

```python
# Заполняем отсутствующие значения 
data['1 star ratings'] = data['1 star ratings'].replace(0,np.nan)
data['1 star ratings'] = data['1 star ratings'].fillna(data['1 star ratings'].mean())
```
